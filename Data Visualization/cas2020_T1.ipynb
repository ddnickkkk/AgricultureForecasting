```python
df = pd.read_csv("cas2020_T1.csv")
df
```

```python
gp_pro = df.groupby('Province')

female = gp_pro['Female'].sum()
male = gp_pro['Male'].sum()
total = gp_pro['Total'].sum()

df = pd.DataFrame({'Female': female,
                    'Male': male,
                    'Total': total
              })
df
```

```python
# Get the number of rows in the dataframe
num_rows = len(df)

# the indices of rows to be deleted
delete_indices = [i for i in range(1, num_rows, 3)] + [i for i in range(2, num_rows, 3)]

# delete every 2 rows from each column
df = df.drop(df.index[delete_indices])
df
```

```python
p = []
for i in df['province']:
    p.append(i)
```

```python
f = []
for j in df['female']:
    f.append(j)
```

```python
m = []
for k in df['male']:
    m.append(k)
```

```python
t = []
for l in df['total']:
    t.append(l)
```

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# Set position of bar on X axis
br1 = np.arange(len(f))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, f, color ='r', width = barWidth,
		edgecolor ='grey', label ='Female')
plt.bar(br2, m, color ='g', width = barWidth,
		edgecolor ='grey', label ='Male')
plt.bar(br3, t, color ='b', width = barWidth,
		edgecolor ='grey', label ='Total')

# Adding Xticks
plt.xlabel('Province', fontweight ='bold', fontsize = 15)
plt.ylabel('Number of each gender', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(p))],
		p)
plt.title("Number of Female vs Male of each Province")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(p, f)

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
	ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

# Add x, y gridlines
ax.grid(b = True, color ='grey',
		linestyle ='-.', linewidth = 0.5,
		alpha = 0.2)

# Show top values
ax.invert_yaxis()

# Add annotation to bars
for i in ax.patches:
	plt.text(i.get_width()+0.2, i.get_y()+0.5,
			str(round((i.get_width()), 2)),
			fontsize = 10, fontweight ='bold',
			color ='grey')

# Add Plot Title
ax.set_title('Female with Agriculture Field in Cambodia',
			loc ='left', )

# Add Text watermark
fig.text(0.9, 0.15, 'Agricultural in Cambodia', fontsize = 12,
		color ='grey', ha ='right', va ='bottom',
		alpha = 0.7)

# Show Plot
plt.show()
```

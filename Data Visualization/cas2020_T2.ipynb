```python
# to_delete1 = 'ASIKHM024: '
# p1 = df['INDICATOR: Indicator'].str.replace(to_delete1,"")
# p1

# to remove ASIKHM in province-names
pro_code = ["","007", "018", "094", "005", "011", "016", "091", "008", "014", "021","092", "002", "006","017","024",
           "093","009","023","010","013","019","003","012","020","025","001","004","015","022"]
for code in pro_code:
    to_delete = f'ASIKHM{code}: '
    df['INDICATOR: Indicator'] = df['INDICATOR: Indicator'].str.replace(to_delete, "")

# to remove code in crops
c_code = ["_T","101","102","103","105","203","302","402","403","441","522","811","801",
            "816","818","861","911","921","931","202","401","502","521","524","527",
            "530","532","562","817","820","828"]
for code in c_code:
    to_delete = f'{code}: '
    df['REF_AREA: Reference area'] = df['REF_AREA: Reference area'].str.replace(to_delete, "")

# to drop some columns
df = df.drop(['RESPONSIBLE_AGENCY: Responsible agency', 'DATA_SOURCE: Data source', 
            'INDICATOR_DESCRIPTION: Indicator description', 'CROP: Crop', 
            'FREQ: Frequency', 'UNIT_MEASURE: Unit of measure', 'OBS_VALUE: Observation value', 
            'Unnamed: 10'], axis=1)
df
```

There are 31 types of crops, so I would like to check each type according to each province in Cambodia.

```python
# Non-aromatic paddy
paddy = df[df["REF_AREA: Reference area"] == "Non-aromatic paddy"]

bar_name = paddy["INDICATOR: Indicator"]
plt.bar(bar_name, paddy["TIME_PERIOD: Time period"])
plt.xlabel("Province")
plt.ylabel("Amount of crop(Non-aromatic paddy)")
plt.title("Number of crop Non-aromatic paddy in different provinces")
plt.xticks(rotation=90)
plt.show()
```
